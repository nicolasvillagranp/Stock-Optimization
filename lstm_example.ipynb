```python
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import clear_output
import os
import time
```

#### Plot de los resultados obtenidos por LSTM

```python
# Get the names of all the files stored in a folder
def get_file_names(folder):
    return os.listdir(folder)
names = get_file_names('LSTM\data_pred')
for i in names:
    df = pd.read_csv('LSTM\data_pred\\'+i)
    # Plot actual and predictions
    plt.figure(figsize=(10, 5))
    plt.plot(df['Actual'], label='Actual')
    plt.plot(df['Prediction'], label='Predictions')
    plt.title(f'Actual vs Predicted for {i[:-4]}')
    plt.legend()
    plt.show()
    time.sleep(2)
    clear_output(wait=True)
```
