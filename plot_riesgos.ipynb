```python
import matplotlib.pyplot as plt
import seaborn as sns
import json
import numpy as np
import scipy.stats as stats
```

#### Cálculo de la distribución del riesgo para elegir un threshold adecuado

```python
# open std json
with open('std.json') as f:
    data = json.load(f)
    # Sample list of numbers (replace this with your own data)
    data = list(data.values())
    # Create a combined histogram and kernel density estimate plot using seaborn
    sns.set_style("white")  # Set the style of the plot
    sns.histplot(data, kde=True, bins=12, stat="density", common_norm=False)
    plt.axvline(np.mean(data)+2*np.std(data), color='r', linewidth=1, label = 'mean + 2 std')
    # Plot a red fill until the last line
    plt.axvspan( min(data),np.mean(data)+2*np.std(data), color='r', alpha=0.2)
    plt.xlabel("Values")
    plt.ylabel("Density")
    plt.title("Histogram with Density of risk")
    plt.legend()
    # Show the plot
    plt.show()
```
